```python
def no_more(funding):
    funding = funding.replace('$', '').strip()
    if 'B' in funding:
        return float(funding.replace('B', ''))
    elif 'M' in funding:
        return float(funding.replace('M', ''))/1000
    

df['Funding (In Billion)'] = df['Funding (In Billion)'].apply(no_more)
```

```python
df_usa_2021['Industry'] = df_usa_2021['Industry'].replace('Artificial intelligence', 'Artificial Intelligence')
```

After grouping the companies by their respective industry and summing up the valuation of the company, we found out that there was an error which was caused due to "Artificial Intelligence" , and "Artificial intelligence" being two different kind of industry whereas both terms mean the same, so we will try to rectify the error. i have thoroughly checked if this will affect our EDA above this step, It doesn't as we have not used Industry column to our visuals.

```python
df_usa_city = df_usa_city [['Industry', 'Valuation']].sort_values(by='Valuation', ascending = False)
plt.figure(figsize=(18,10))
bars = plt.bar(df_usa_city['Industry'], df_usa_city['Valuation'], color='blue')
plt.title('Valuation of Companies by Industry In USA in 2021')
plt.ylabel('Valuation (in Billion)')
plt.xlabel('Industry')
plt.xticks(rotation=60)

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        str(int(bar.get_height())),
        ha='center', 
        va = 'bottom'
    )
plt.show()
```

After looking at the visualisation, we can clearly see that, during the year 2021, which was the year after the Covid where most of the countries had all their business shut down physically and entered digitalisation with their work.

Fintech, Internet software and services, Health are the top '3' industries that grew exponentially and had most unicorn companies during the year 2021. Finance paired with Tech boomed, so did the Software companies, and due to the outbreak, Health industry was profitable as well as struggling to fulfill the demand at the time.
