```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
df['Day_Month'] = df['Date'].dt.strftime('%B')
```

```python
df.drop(['Date'], axis = 1 , inplace = True)
```

```python
df.head()
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
```

```python
df['Day_Month'] = pd.Categorical(df['Day_Month'], categories=month_order, ordered=True)
```

```python
df = df.sort_values('Day_Month').reset_index(drop=True)
```

```python
df.head()
```

```python
 rating_mean = df.groupby('Product line')['Rating'].mean()
```

```python
rating_mean
```

```python
monthly_sales = df.groupby('Day_Month', observed = True)['Total'].sum().reset_index()
```

```python
number_of_months = monthly_sales['Day_Month'].nunique()
print(f"Number of unique months: {number_of_months}")
#Checking the number of months in the data frame, and it seems there are only 3 months, January , February, and March.
```

```python
monthly_sales
```

```python
monthly_profit = df.groupby('Day_Month', observed = True)['gross income'].sum().reset_index()
```

```python
monthly_profit
```

```python
plt.bar(monthly_profit['Day_Month'], monthly_profit['gross income'])
plt.xlabel('Months', loc = 'left')
plt.ylabel('Gross Income', loc ='center')
plt.title('Gross Income for the Months')
plt.show()
```

```python
products_sold = df.groupby(['Day_Month','Product line'], observed = True)['Quantity'].sum().reset_index()
```

```python
products_sold
```

```python
products_sold_pivot = products_sold.pivot(index='Day_Month', columns='Product line', values='Quantity')
```

```python
products_sold_pivot
```

```python
plt.figure(figsize=(12,6))
for product_line in products_sold_pivot.columns:
    plt.plot(products_sold_pivot.index, products_sold_pivot[product_line], marker = 'o', label = product_line)

plt.title('Product Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.legend(title='Product Line')
plt.tight_layout()
plt.show()
    
```

```python
profit_by_member = df.groupby('Customer type')['gross income'].sum().reset_index()
```

```python
profit_by_member
```

```python
plt.figure(figsize=(6,5))
plt.pie(profit_by_member['gross income'] , labels = profit_by_member['Customer type'], autopct= '%1.2f%%', startangle= 90)
plt.title('Revenue Distribution by Gender')
plt.legend(title = "Member Type" , loc = "best")
```

```python
revenue_by_gender = df.groupby('Gender')['gross income'].sum().reset_index()
```

```python
revenue_by_gender
```

```python
plt.figure(figsize=(8,6))
plt.pie(revenue_by_gender['gross income'],labels = revenue_by_gender['Gender'], autopct='%1.2f%%' , startangle = 90)
plt.title('Revenue Distribution by Gender')
plt.legend(title="Gender", loc="best")
plt.show()
```
